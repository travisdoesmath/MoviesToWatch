```python
import pymongo
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
conn = "mongodb://localhost:27017"
client = pymongo.MongoClient(conn)
db = client.movies_db
```

```python
movies = db.movies.find()
```

```python
df = pd.DataFrame([x for x in movies])
```

```python
df
```

```python
# fix the year data
df[df['year'].apply(lambda x: x[-4:] == 'ovie')]
```

```python
df.loc[425,'year'] = '1965'
```

```python
df['year'] = df['year'].apply(lambda x: int(x[-4:]))
```

```python
df
```

```python
df['decade'] = df['decade'].apply(lambda x: int(10*np.floor(x / 10)))
df['count'] = 1
```

```python
seen_by_decade = df.groupby('decade').sum()[['seen','count']]
seen_by_decade['percent'] = seen_by_decade['seen'] / seen_by_decade['count']
seen_by_decade[['seen','count']].plot()
plt.show()
seen_by_decade['percent'].plot.bar()
plt.show()
```
